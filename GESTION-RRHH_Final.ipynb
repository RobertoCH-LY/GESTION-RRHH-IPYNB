```python
from tkinter import *
from tkinter import scrolledtext
import sqlite3
import re 
from datetime import datetime

# ==============================================================================
# 1. PANTALLA DE REGISTRO (ALTAS)
# ==============================================================================
def abrir_ventana_altas():
    ventana = Toplevel()
    ventana.title("Pantalla Altas")
    ventana.resizable(0,0)
    ventana.config(bg="#f8f8f8")

    # Variables
    v_nombre = StringVar()
    v_inicio = StringVar()
    v_nacimiento = StringVar()
    v_direccion = StringVar()
    v_nif = StringVar()
    v_naf = StringVar()
    v_genero = StringVar()
    v_departamento = StringVar()
    v_puesto = StringVar()
    v_telefono = StringVar()
    v_salario = StringVar()
    v_email = StringVar()
    v_pagas = StringVar()
    v_mensaje = StringVar()

    # --- DISEÑO ---
    Label(ventana, text="APELLIDOS Y NOMBRE", font=("Arial", 10, "bold"), width=25, anchor="center", bg="#f8f8f8").grid(
        row=0, column=0, columnspan=4, padx=20, pady=(15,5)
    )
    Entry(ventana, justify="center", width=135, textvariable=v_nombre).grid(
        row=1, column=0, columnspan=4, padx=20, pady=10, sticky="w"
    )

    Label(ventana, text="FECHA INICIO", font=("Arial", 10, "bold"), width=15, anchor="center", bg="#f8f8f8").grid(
        row=2, column=0, padx=20, pady=(15,5)
    )
    Entry(ventana, width=28, justify="center", textvariable=v_inicio).grid(
        row=3, column=0, padx=(20,0), pady=5, sticky="w"
    )

    Label(ventana, text="FECHA NACIMIENTO", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=2, column=1, padx=(0,20), pady=(15,5)
    )
    Entry(ventana, width=28, justify="center", textvariable=v_nacimiento).grid(
        row=3, column=1, padx=0, pady=5, sticky="w"
    )

    Label(ventana, text="DIRECCIÓN", font=("Arial", 10, "bold"), width=15, anchor="center", bg="#f8f8f8").grid(
        row=2, column=2, columnspan=2, padx=20, pady=(15,5)
    )
    Entry(ventana, width=68, justify="center", textvariable=v_direccion).grid(
        row=3, column=2, columnspan=2, padx=(0,20), pady=5, sticky="w"
    )

    Label(ventana, text="NIF", font=("Arial", 10, "bold"), width=15, anchor="center", bg="#f8f8f8").grid(
        row=4, column=0, columnspan=2, pady=(15,5)
    )
    Entry(ventana, width=62, justify="center", textvariable=v_nif).grid(
        row=5, column=0, columnspan=2, padx=20, pady=5, sticky="w"
    )

    Label(ventana, text="NÚMERO DE AFILIACIÓN SS", font=("Arial", 10, "bold"), width=25, anchor="center", bg="#f8f8f8").grid(
        row=4, column=2, columnspan=2, pady=(15,5)
    )
    Entry(ventana, justify="center", width=68, textvariable=v_naf).grid(
        row=5, column=2, columnspan=2, pady=5, sticky="w"
    )

    Label(ventana, text="GÉNERO", font=("Arial", 10, "bold"), width=15, anchor="center", bg="#f8f8f8").grid(
        row=6, column=0,pady=(15,5)
    )
    Entry(ventana, width=28,justify="center", textvariable=v_genero).grid(
        row=7, column=0, padx=20, pady=5, sticky="w"
    )

    Label(ventana, text="DEPARTAMENTO", font=("Arial", 10, "bold"), width=15, anchor="center", bg="#f8f8f8").grid(
        row=6, column=1, columnspan=2, pady=(15,5)
    )
    Entry(ventana, width=60, justify="center", textvariable=v_departamento).grid(
        row=7, column=1, columnspan=2, pady=5, sticky="w"
    )

    Label(ventana, text="PUESTO", font=("Arial", 10, "bold"), width=15, anchor="center", bg="#f8f8f8").grid(
        row=6, column=3, pady=(15,5)
    )
    Entry(ventana, width=39, justify="center", textvariable=v_puesto).grid(
        row=7, column=3, pady=5
    )

    Label(ventana, text="TELÉFONO", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=8, column=0, pady=(30,5), padx=(40,5), sticky="w"
    )
    Entry(ventana, width=28, justify="center", textvariable=v_telefono).grid(
        row=8, column=0, pady=(30,5), padx=(10,20), columnspan=2
    )

    Label(ventana, text="SALARIO MENSUAL", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=8, column=1, columnspan=2, pady=(30,5), padx=(15,0)
    )
    Entry(ventana, width=58, justify="center", textvariable=v_salario).grid(
        row=8, column=2, pady=(30,5), padx=(0,20), columnspan=2, sticky="e"
    )

    Label(ventana, text="EMAIL", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=9, column=0, pady=(30,5), padx=(40,5), sticky="w"
    )
    Entry(ventana, width=28, justify="center", textvariable=v_email).grid(
        row=9, column=0, pady=(30,5), padx=(10,20), columnspan=2
    )

    Label(ventana, text="PAGAS EXTRA", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=9, column=1, columnspan=2, pady=(30,5), sticky="w",padx=(115,0)
    )
    Entry(ventana, width=58, justify="center", textvariable=v_pagas).grid(
        row=9, column=2, pady=(30,5), padx=(0,20), columnspan=2, sticky="e"
    )

    Entry(ventana, width=60,justify="center", textvariable=v_mensaje, fg="#2c3e50").grid(
        row=11, column=0, columnspan=3, padx=20, pady=(40,10), sticky="ew", ipady=35
    )

    # --- VALIDACIONES ---
    def validar_nif_nie(documento):
        if re.match(r'^\d{8}[A-Z]$', documento):
            letras = "TRWAGMYFPDXBNJZSQVHLCKE"
            return documento[-1] == letras[int(documento[:-1]) % 23]
        elif re.match(r'^[XYZxyz]\d{7}[A-Za-z]$', documento):
            mapa = {'x':'0', 'y':'1', 'z':'2'}
            num = mapa.get(documento[0].lower()) + documento[1:-1]
            letras = "TRWAGMYFPDXBNJZSQVHLCKE"
            return documento[-1].upper() == letras[int(num) % 23]
        return False

    def validar_naf(naf):
        if re.match(r'^\d{12}$', naf):
            a, b, c = int(naf[0:2]), int(naf[2:10]), int(naf[10:])
            res = (b + a * 10000000) % 97 if b < 10000000 else int(str(a) + str(b)) % 97
            return c == res
        return False

    def guardar_empleado():
        if not validar_nif_nie(v_nif.get()):
            v_mensaje.set("Error: NIF/NIE incorrecto")
            return
        if not validar_naf(v_naf.get()):
            v_mensaje.set("Error: NAF incorrecto")
            return
        
        try:
            conn = sqlite3.connect("empleados.db")
            c = conn.cursor()
            c.execute("""CREATE TABLE IF NOT EXISTS empleados (
                id INTEGER PRIMARY KEY AUTOINCREMENT,apellido_nombre TEXT, fecha_inicio TEXT, fecha_nacimiento TEXT,
                direccion TEXT, nif TEXT, naf TEXT, genero TEXT,
                departamento TEXT, puesto TEXT, telefono TEXT,
                salario_mensual TEXT, email TEXT, pagas_extra TEXT, fecha_fin TEXT
            )""")
            c.execute("""INSERT INTO empleados (
                apellido_nombre, fecha_inicio, fecha_nacimiento, direccion, nif, naf,
                genero, departamento, puesto, telefono, salario_mensual, email, pagas_extra
            ) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?)""", (
                v_nombre.get(), v_inicio.get(), v_nacimiento.get(), v_direccion.get(),
                v_nif.get(), v_naf.get(), v_genero.get(), v_departamento.get(),
                v_puesto.get(), v_telefono.get(), v_salario.get(), v_email.get(), v_pagas.get()
            ))
            conn.commit()
            conn.close()
            v_mensaje.set("Empleado insertado correctamente")
        except Exception as e:
            v_mensaje.set(f"Error: {e}")

    Button(ventana, text="INSERTAR", font=("Segoe UI", 14, "bold"), command=guardar_empleado, bg="#e74c3c", fg="#ffffff", activebackground="#c0392b", width=20).grid(
        row=11, column=3, padx=20, pady=(40,10)
    )


# ==============================================================================
# 2. PANTALLA INFORME
# ==============================================================================
def abrir_ventana_informe():
    ventana2 = Toplevel()
    ventana2.title("Pantalla Informe")
    ventana2.resizable(0,0)
    ventana2.config(bg="#f8f8f8")
    
    # Variables
    v_empleados = StringVar()
    v_media_edad = StringVar()
    v_retribucion = StringVar()
    v_porc_mujeres = StringVar()
    v_edad_mujeres = StringVar()
    v_retri_mujeres = StringVar()
    v_porc_hombres = StringVar()
    v_edad_hombres = StringVar()
    v_retri_hombres = StringVar()

    # Diseño
    Label(ventana2, text="EMPLEADOS", font=("Arial", 10, "bold"),width=20, anchor="center", bg="#f8f8f8").grid(
        row=0, column=3, columnspan=2, pady=(30,5),padx=(40), sticky="e"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_empleados).grid(
        row=1, column=3, columnspan=2, padx=(90,50), pady=(15,20), ipady=30
    )
    Label(ventana2, text="MEDIA EDADES", font=("Arial", 10, "bold"),width=15, anchor="center", bg="#f8f8f8").grid(
        row=0, column=5, columnspan=2, pady=(30,5),padx=(57), sticky="w"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_media_edad).grid(
        row=1, column=5, columnspan=2, padx=50, pady=(15,20), ipady=30
    )
    Label(ventana2, text="RETRIBUCIÓN MEDIA", font=("Arial", 10, "bold"),width=20, anchor="center", bg="#f8f8f8").grid(
        row=0, column=7, columnspan=2, pady=(30,5), padx=(48), sticky="w"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_retribucion).grid(
        row=1, column=7, columnspan=2, padx=(50,80), pady=(15,20), ipady=30
    )
    Label(ventana2, text="% MUJERES", font=("Arial", 10, "bold"),width=20, anchor="center", bg="#f8f8f8").grid(
        row=2, column=3, columnspan=2, pady=(30,5),padx=(40), sticky="e"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_porc_mujeres).grid(
        row=3, column=3, columnspan=2, padx=(90,50), pady=(15,20), ipady=30
    )
    Label(ventana2, text="MUJERES", font=("Arial", 10, "bold"),width=15, anchor="center", bg="#f8f8f8").grid(
        row=2, column=5, columnspan=2, pady=(30,5),padx=(57), sticky="w"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_edad_mujeres).grid(
        row=3, column=5, columnspan=2, padx=50, pady=(15,20), ipady=30
    )
    Label(ventana2, text="MUJERES", font=("Arial", 10, "bold"),width=20, anchor="center", bg="#f8f8f8").grid(
        row=2, column=7, columnspan=2, pady=(30,5), padx=(48), sticky="w"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_retri_mujeres).grid(
        row=3, column=7, columnspan=2, padx=(50,80), pady=(15,20), ipady=30
    )
    Label(ventana2, text="% HOMBRES", font=("Arial", 10, "bold"),width=20, anchor="center", bg="#f8f8f8").grid(
        row=4, column=3, columnspan=2, pady=(30,5),padx=(40), sticky="e"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_porc_hombres).grid(
        row=5, column=3, columnspan=2, padx=(90,50), pady=(15,20), ipady=30
    )
    Label(ventana2, text="HOMBRES", font=("Arial", 10, "bold"),width=15, anchor="center", bg="#f8f8f8").grid(
        row=4, column=5, columnspan=2, pady=(30,5),padx=(57), sticky="w"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_edad_hombres).grid(
        row=5, column=5, columnspan=2, padx=50, pady=(15,20), ipady=30
    )
    Label(ventana2, text="HOMBRES", font=("Arial", 10, "bold"),width=20, anchor="center", bg="#f8f8f8").grid(
        row=4, column=7, columnspan=2, pady=(30,5), padx=(48), sticky="w"
    )
    Entry(ventana2, width=25, justify="center", textvariable=v_retri_hombres).grid(
        row=5, column=7, columnspan=2, padx=(50,80), pady=(15,20), ipady=30
    )

    def calcular_estadisticas():
        conn = sqlite3.connect("empleados.db")
        cur = conn.cursor()
        try:
            cur.execute("SELECT fecha_nacimiento, salario_mensual, genero FROM empleados")
        except sqlite3.OperationalError:
            cur.execute("ALTER TABLE empleados ADD COLUMN fecha_nacimiento TEXT")
            conn.commit()
            cur.execute("SELECT fecha_nacimiento, salario_mensual, genero FROM empleados")
        datos = cur.fetchall()
        cur.execute("SELECT COUNT(*) FROM empleados")
        total_emp = cur.fetchone()[0]
        v_empleados.set(total_emp)
        if not datos: 
            conn.close()
            return
        l_edades, l_salarios, l_edades_m, l_salarios_m, l_edades_h, l_salarios_h = [], [], [], [], [], []
        anio_actual = datetime.now().year
        for fecha, salario, genero in datos:
            edad = 0
            if fecha and "/" in fecha:
                try: 
                    anio = int(fecha.split("/")[-1])
                    edad = anio_actual - anio
                except: pass
            sueldo = float(salario) if salario else 0
            l_edades.append(edad)
            l_salarios.append(sueldo)
            if genero and genero.lower().strip() == "femenino": 
                l_edades_m.append(edad)
                l_salarios_m.append(sueldo)
            elif genero and genero.lower().strip() == "masculino": 
                l_edades_h.append(edad)
                l_salarios_h.append(sueldo)
        def media(lista): return round(sum(lista)/len(lista), 2) if lista else 0
        v_media_edad.set(media(l_edades))
        v_retribucion.set(media(l_salarios))
        v_edad_mujeres.set(media(l_edades_m))
        v_retri_mujeres.set(media(l_salarios_m))
        v_porc_mujeres.set(f"{round((len(l_edades_m)/total_emp)*100,2)}%" if total_emp else "0%")
        v_edad_hombres.set(media(l_edades_h))
        v_retri_hombres.set(media(l_salarios_h))
        v_porc_hombres.set(f"{round((len(l_edades_h)/total_emp)*100,2)}%" if total_emp else "0%")
        conn.close()
    
    calcular_estadisticas()


# ==============================================================================
# 3. PANTALLA FICHERO (MOSTRAR EN PANTALLA CON SCROLL)
# ==============================================================================
def abrir_ventana_fichero():
    ventana3 = Toplevel()
    ventana3.title("Visualización de Fichas")
    ventana3.resizable(0,0)
    ventana3.config(bg="#f8f8f8")
    
    Label(ventana3, text="LISTADO DE EMPLEADOS", font=("Arial", 14, "bold"), fg="#2c3e50", bg="#f8f8f8").pack(
        pady=(20, 10)
    )
    
    area_texto = scrolledtext.ScrolledText(ventana3, width=90, height=35, font=("Consolas", 10)) 
    area_texto.pack(padx=20, pady=10)

    def mostrar_datos():
        conn = sqlite3.connect("empleados.db")
        cur = conn.cursor()
        cur.execute("SELECT * FROM empleados")
        filas = cur.fetchall()
        cols = [d[0] for d in cur.description]
        conn.close()

        area_texto.config(state='normal')
        area_texto.delete(1.0, END)

        if not filas:
            area_texto.insert(INSERT, "No hay empleados registrados en la base de datos.")
        else:
            for r in filas:
                data = {cols[i]: r[i] if r[i] else "" for i in range(len(cols))}
                s_mes = float(data.get('salario_mensual') or 0)
                pagas = float(data.get('pagas_extra') or 0)
                s_anual = s_mes * pagas
                prorrata = (s_mes * (pagas - 12)) / 12 if pagas > 12 else 0
                s_mes_str = f"{s_mes:.2f}".replace('.', ',')
                s_anual_str = f"{s_anual:.2f}".replace('.', ',')
                prorrata_str = f"{prorrata:.2f}".replace('.', ',')
                
                ficha = ""
                ficha += f"{data.get('apellido_nombre', '').upper()}\n"
                ficha += f"{data.get('direccion', '').upper()}\n\n"
                ficha += f"NIF {data.get('nif')}".ljust(35) + f"NAF {data.get('naf')}\n"
                fecha_baja = data.get('fecha_fin') if data.get('fecha_fin') else ""
                ficha += f"F. ALTA: {data.get('fecha_inicio')}".ljust(35) + f"F. BAJA: {fecha_baja}\n\n"
                ficha += "CONCEPTOS SALARIALES\n\n"
                ficha += f"SALARIO MENSUAL   {s_mes_str}".ljust(45) + f"SALARIO ANUAL     {s_anual_str}\n"
                ficha += f"PRORRATA DE PAGAS {prorrata_str}".ljust(45) + f"NÚMERO PAGAS      {int(pagas)}\n"
                ficha += "-" * 100 + "\n\n"
                area_texto.insert(INSERT, ficha)

        area_texto.config(state='disabled')
        
    Button(ventana3, text="ACTUALIZAR DATOS", bg="#e74c3c", fg="#ffffff", font=("Segoe UI", 14, "bold"), activebackground="#c0392b", command=mostrar_datos).pack(
        pady=10
    )
    
    mostrar_datos()


# ==============================================================================
# 4. PANTALLA GESTION BAJAS
# ==============================================================================
def abrir_ventana_bajas():
    ventana4 = Toplevel()
    ventana4.title("Pantalla Bajas")
    ventana4.resizable(0,0)
    ventana4.config(bg="#f8f8f8")
    
    # Variables
    v_id = StringVar()
    v_fecha_baja = StringVar()
    v_inicio = StringVar()
    v_fin = StringVar()
    v_nombre = StringVar()
    v_nif = StringVar()
    v_naf = StringVar()
    v_salario_mes = StringVar()
    v_num_pagas = StringVar()
    v_salario_anual = StringVar()
    v_prorrata = StringVar()
    v_mensaje = StringVar()

    # Diseño
    Label(ventana4, text="CÓDIGO EMPLEADO", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=0, column=0, padx=(20), pady=(35,20)
    )
    Entry(ventana4, width=23, justify="center", textvariable=v_id).grid(
        row=0, column=0, columnspan=2, pady=(35,20),padx=(130,0)
    )
    Label(ventana4, text="FECHA BAJA", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=0, column=1, columnspan=2, padx=(0,310), pady=(35,20) 
    )
    Entry(ventana4, width=23, justify="center", textvariable=v_fecha_baja).grid(
        row=0, column=2, columnspan=2, pady=(35,20),padx=90,sticky="w"
    )
    Label(ventana4, text="FECHA INICIO", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=2, column=0, padx=20, pady=(20,5)
    )
    Entry(ventana4, width=27, justify="center", textvariable=v_inicio).grid(
        row=3, column=0, padx=20, pady=(5,20)
    )
    Label(ventana4, text="FECHA FIN", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=2, column=1, padx=20, pady=(20,5)
    )
    Entry(ventana4, width=27, justify="center", textvariable=v_fin).grid(
        row=3, column=1, padx=10, pady=(5,20)
    )
    Label(ventana4, text="APELLIDOS Y NOMBRE", font=("Arial", 10, "bold"),width=23, anchor="center", bg="#f8f8f8").grid(
        row=2, column=2, columnspan=2, padx=(20,90), pady=(20,5)
    )
    Entry(ventana4, width=64, justify="center", textvariable=v_nombre).grid(
        row=3, column=2, columnspan=2, padx=10, pady=(5,20), sticky="w"
    )
    Label(ventana4, text="NIF", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=4, column=0, columnspan=2, padx=20, pady=(20,5)
    )
    Label(ventana4, text="NÚMERO DE AFILIACIÓN SS", font=("Arial", 10, "bold"),width=23, anchor="center", bg="#f8f8f8").grid(
        row=4, column=2, columnspan=2, padx=(20,90), pady=(20,5)
    )
    Entry(ventana4, width=60, justify="center", textvariable=v_nif).grid(
        row=5, column=0, columnspan=2, padx=20, pady=(5,20), sticky="w"
    )
    Entry(ventana4, width=64, justify="center", textvariable=v_naf).grid(
        row=5, column=2, columnspan=2, padx=(12,20), pady=(5,20), sticky="w"
    )
    Label(ventana4, text="SALARIO BRUTO MES", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=6, column=0, padx=20, pady=(30,5), sticky="w"
    )
    Entry(ventana4, width=35, justify="center", textvariable=v_salario_mes).grid(
        row=6, column=0, columnspan=2, padx=(0,25), pady=(30,5), sticky="e"
    )
    Label(ventana4, text="NÚMERO PAGAS", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=6, column=2, padx=20, pady=(20,5), sticky="w"
    )
    Entry(ventana4, width=43, justify="center", textvariable=v_num_pagas).grid(
        row=6, column=2, columnspan=1, padx=134, pady=(20,5)
    )
    Label(ventana4, text="SALARIO ANUAL", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=7, column=0, columnspan=2, padx=20, pady=(0,5), sticky="w"
    )
    Entry(ventana4, width=35, justify="center", textvariable=v_salario_anual).grid(
        row=7, column=0, columnspan=2, padx=(0,25), pady=(0,5),sticky="e"
    )
    Label(ventana4, text="PRORRATA PAGAS", font=("Arial", 10, "bold"), bg="#f8f8f8").grid(
        row=7, column=0, columnspan=2, padx=20, pady=(80,5), sticky="w"
    )
    Entry(ventana4, width=35, justify="center", textvariable=v_prorrata).grid(
        row=7, column=0, columnspan=2, padx=(0,25), pady=(80,5),sticky="e"
    )
    Entry(ventana4, justify="center", font=("Arial", 10, "bold"), width=53, textvariable=v_mensaje).grid(
        row=7, column=2, padx=20,pady=(25,0), ipady=20, sticky="w"
    )

    # Funciones internas
    def cargar_empleado():
        conn = sqlite3.connect("empleados.db")
        c = conn.cursor()
        c.execute("SELECT apellido_nombre, nif, naf, salario_mensual, pagas_extra, fecha_inicio, fecha_fin FROM empleados WHERE id = ?", (v_id.get(),))
        res = c.fetchone()
        conn.close()
        if res:
            v_nombre.set(res[0])
            v_nif.set(res[1])
            v_naf.set(res[2])
            s = float(res[3]) if res[3] else 0
            p = int(res[4]) if res[4] else 0
            v_salario_mes.set(str(s))
            v_num_pagas.set(str(p))
            v_salario_anual.set(str(round(s * p, 2)))
            v_prorrata.set(str(round((s * (p - 12)) / 12, 2)) if p > 12 else "0")
            v_inicio.set(res[5])
            v_fin.set(res[6])
            v_mensaje.set("Datos cargados correctamente")
        else: 
            v_mensaje.set("Empleado no encontrado")

    def insertar_baja():
        conn = sqlite3.connect("empleados.db")
        c = conn.cursor()
        c.execute("UPDATE empleados SET fecha_fin = ? WHERE id = ?", (v_fecha_baja.get(), v_id.get()))
        conn.commit()
        conn.close()
        cargar_empleado()
        v_mensaje.set("Fecha de baja insertada correctamente")
        
    Button(ventana4, text="INSERTAR FECHA BAJA", font=("Segoe UI", 14, "bold"), bg="#e74c3c", fg="#ffffff", activebackground="#c0392b", width=40, command=insertar_baja).grid(
        row=11, column=0, columnspan=2, padx=(45,0), pady=(50,20), ipady=20
    )
    Button(ventana4, text="CARGA EMPLEADO", font=("Segoe UI", 14, "bold"), bg="#e74c3c", fg="#ffffff", activebackground="#c0392b", width=40, command=cargar_empleado).grid(
        row=11, column=2, columnspan=2, padx=(20,40), pady=(48,20), ipady=20, sticky="w"
    )


# ==============================================================================
# MENÚ PRINCIPAL DEL SISTEMA
# ==============================================================================

principal = Tk();principal.title("GESTOR EMPLEADOS");principal.resizable(0,0);principal.config(bg="#f8f8f8")
principal.iconbitmap('seta.ico')

Label(principal, text="GESTOR DE EMPLEADOS", bg="#f8f8f8", font=("Impact", 30), fg="#2c3e50").grid(row=0, column=4, columnspan=4, pady=(40,20))
imagen_central = PhotoImage(file="seta.png")

Label(principal, image=imagen_central, bg="#f8f8f8").grid(row=2, column=4, columnspan=4, pady=(10, 30))
Button(principal, text="ALTAS", bg="#e74c3c", fg="#ffffff", font=("Segoe UI", 14, "bold"), width=15, height=1, activebackground="#c0392b", command=abrir_ventana_altas).grid(row=4, column=3, columnspan=2, padx=(200,10), pady=10, sticky="ew")
Button(principal, text="GESTIÓN BAJAS", bg="#e74c3c", fg="#ffffff", font=("Segoe UI", 14, "bold"), width=15, height=1, activebackground="#c0392b", command=abrir_ventana_bajas).grid(row=4, column=5, columnspan=2, padx=10, pady=10, sticky="ew")
Button(principal, text="FICHERO", bg="#e74c3c", fg="#ffffff", font=("Segoe UI", 14, "bold"), width=15, height=1, activebackground="#c0392b", command=abrir_ventana_fichero).grid(row=5, column=3, columnspan=2, padx=(200,10), pady=10, sticky="ew")
Button(principal, text="INFORME", bg="#e74c3c", fg="#ffffff", font=("Segoe UI", 14, "bold"), width=15, height=1, activebackground="#c0392b", command=abrir_ventana_informe).grid(row=5, column=5, columnspan=2, padx=10, pady=10, sticky="ew")

principal.mainloop()
```
